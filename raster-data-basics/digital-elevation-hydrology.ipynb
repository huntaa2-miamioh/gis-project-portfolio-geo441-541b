2. **Finally, in a brief paragraph explain why this data layer is an appropriate choice for your project topic**, including a discussion of the appropriateness of the spatial and temporal resolutions (i.e. the granularity/scale of the dataset and its date of collection).
<mark> The NASA Black Marble (VNP46A4) Nighttime Lights dataset is an appropriate choice for this project because it provides a high-resolution, consistent measure of human activity and economic intensity that directly relates to commuting and work-from-home trends. As office areas become less occupied and commercial activity decreases during widespread remote work, nighttime light intensity often diminishes in those locations, making this dataset a decent proxy for shifts in workplace presence and urban vitality. Its spatial extent of approximately 500 meters allows for detailed analysis across California’s urban and suburban regions, capturing localized changes in brightness associated with return-to-office patterns. The annual temporal resolution is also well suited for comparing pre-pandemic (2019), mid-pandemic (2021), and post-pandemic (2023) periods, providing a consistent timeline to observe long-term patterns rather than short-term fluctuations. Overall, the Black Marble dataset’s scale, temporal coverage, and physical basis in observable light emissions make it an effective raster source for studying changes in commuting-related activity. </mark>
