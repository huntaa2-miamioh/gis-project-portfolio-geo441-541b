## **Part 3: Reflection**

*In a brief paragraph, compare the two maps you created. Describe the visual and data model differences between the DEM datasets visualized in the topology map and the zonal statistics calculated and appended to the parcel dataset in the second map.*

<mark> The first map in the created (Oxford Topography) visualizes the Digital Evelvation Model (DEM) data as a continuous raster surface, symbolized by elevation values and enhanced with hillshade to show terrain shape and topographic variations. Each raster cell stores an elevation value, so the data model is pixel-based and represents the landscape as a smooth, uninterrupted surface. In contrast, the second map takes the DEM-derived slope raster and summarizes it using zonal statistics, attaching the mean slope value to each parcel in the vector parcel dataset. Instead of displaying continuous terrain, the second map shows distinct polygon features (land parcels), each filled according to the average slope calculated within its boundaries. This shifts the data model from raster (continuous elevation data) to vector (discrete parcels with attribute fields), communicates slope characteristics per parcel rather than raw terrain elevation. </mark>
