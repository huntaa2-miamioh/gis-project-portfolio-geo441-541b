  - Data import and management
    - In my first experience with ArcGIS Pro, I found data import to be relatively straightforward, especially when working out of the data file from Canvas. I like the content bar on the lefthand side that allows users to overview all the data they are managing. I just need to adopt a better system for tracking my data, as I kept losing its location in my drive and getting it mistaken with other files. At one point, I accidentally mistook the add folder button for the way to add data, so now I won't forget how to add data. Overall, I was able to successfully import, organize, and view multiple layers of spatial data and forsee myself getting more familiar pretty quickly.
  - Map creation and annotation
    - Creating maps in ArcGIS Pro was an enjoyable part of the process. I was able to add layers, adjust symbology, and set up legends and scale bars with relative ease. The layout view allowed me to create a professional-looking map with titles and annotations, although it took some time to figure out how to properly format and position text elements. I hypothesize that this will be one of my favorite parts of using GIS is formatting the data in a clear, concise and aesthetically pleasing manner. I appreciated the level of customization available, possibilities are endless!
  - Software interface navigation
    - Navigating the ArcGIS Pro interface was a bit challenging at first due to the number of tools and panels available. It took me a while to understand the difference between the map view and the layout view, and where to find specific geoprocessing tools. However, the ribbon-style interface was familiar from other software like Microsoft Office, which helped. Completing two maps before working on the one that gets submitted made it easier to locate the tools I needed. I still occasionally find myself clicking through multiple tabs to find the right function, but overall I'm gaining confidence. The videos included in the lab were super useful and I plan to refer to those anytime I get stuck.
