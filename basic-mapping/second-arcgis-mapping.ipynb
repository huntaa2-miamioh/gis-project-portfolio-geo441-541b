## **Reflection on data models and management in ArcGIS Pro**


Within ArcGis, lots of data is stored and organized as geodatabases. A geodatabase refers to a central, overall container for spatial and non-spatial data, stored as a cluster of files. This appears as a directory or folder in which you can explore other files. Within this geodatabase, there can be feature datasets which act as a subgroup used to store related spatial data, which can take the form of feature classes that must share a common spatial reference or layer of data. Feature classes can take the form of points, lines, polygons, etc; they must be collections of similar geographic features.
